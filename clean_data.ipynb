```python
import pandas as pd

df = pd.read_csv('tiktok_shop_products.csv')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_replace = df.copy()

df_replace['Rating'] = df['Rating'].astype(float)
df_replace['Original_Price'] = df['Original_Price'].fillna('0')

words_to_remove = [".", "₫", "-", "%"]

for word in words_to_remove:
    df_replace['Sold'] = df_replace['Sold'].str.replace(word, '')
    df_replace['Original_Price'] = df_replace['Original_Price'].str.replace(word, '')
    df_replace['Current_Price'] = df_replace['Current_Price'].str.replace(word, '')
    df_replace['Discount_Percent'] = df_replace['Discount_Percent'].str.replace(word, '')

df_replace['Sold'] = df_replace['Sold'].str.replace('K', '00').astype(int)
df_replace['Original_Price'] = df_replace['Original_Price'].astype(int)
df_replace['Current_Price'] = df_replace['Current_Price'].astype(int)
df_replace['Discount_Percent'] = df_replace['Discount_Percent'].astype(int)

df_replace.head()
```

```python
df_replace.info()
```

```python
df_replace.to_csv('tiktok_shop_products_cleaned.csv', index=False)
```
